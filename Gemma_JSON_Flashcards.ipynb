```python
import google.generativeai as genai
from google.colab import userdata
import json

# 1. API Configuration
genai.configure(api_key=userdata.get('GOOGLE_API_KEY'))

# 2. Using the working model
model = genai.GenerativeModel('models/gemini-2.5-flash')

# 3. JSON Output Prompt
def generate_flashcards(topic):
    prompt = f"""
    You are an expert educational AI. Create 3 advanced quiz questions about "{topic}".
    You MUST output ONLY valid JSON in the exact format below, with no markdown formatting or backticks.

    [
      {{"question": "...", "answer": "...", "difficulty": "Hard"}},
      {{"question": "...", "answer": "...", "difficulty": "Medium"}}
    ]
    """
    response = model.generate_content(prompt)
    return response.text

# 4. Execution
topic_input = input("Enter a tech topic (e.g., MongoDB, Vector Databases, Python Decorators): ")
print("\n--- GENERATING JSON FLASHCARDS ---\n")
json_output = generate_flashcards(topic_input)
print(json_output)
```
